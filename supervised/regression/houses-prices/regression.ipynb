```python
f, (ax1,ax2,ax3,ax4) = plt.subplots(1,4,sharey=True,figsize=(15,4))
ax1.scatter(data_cleaned['LotArea'],data_cleaned['SalePrice'])
ax1.set_title('Price and LotArea')
ax2.scatter(data_cleaned['YearBuilt'],data_cleaned['SalePrice'])
ax2.set_title('Price and YearBuilt')
ax3.scatter(data_cleaned['TotalBsmtSF'],data_cleaned['SalePrice'])
ax3.set_title('Price and TotalBsmtSF')
ax4.scatter(data_cleaned['GarageArea'],data_cleaned['SalePrice'])
ax4.set_title('Price and GarageArea')
plt.show()
```

```python
data_cleaned = data_cleaned.drop(['LotArea'],axis=1)
```

```python
data_cleaned.columns.values
```

```python
cols = ['SalePrice','YearBuilt','TotalBsmtSF','GarageArea']
```

```python
data_preprocessed = data_cleaned[cols]
```

```python
targets = data_preprocessed['SalePrice']
inputs = data_preprocessed.drop(['SalePrice'],axis=1)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs, targets, test_size = 0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat, alpha = 0.2)
```

```python
reg.score(x_train,y_train)
```

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test,y_hat_test,alpha=0.2)
```

```python
sns.distplot(y_train - y_hat)
```

```python
sns.distplot(y_test - y_hat_test)
```
